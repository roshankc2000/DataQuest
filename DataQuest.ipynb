```python
import pandas as pd
import numpy as np
```

```python
battle = pd.read_csv("battles.csv")
```

```python
battle.head()
```

```python
## to know about no of rows & columns
battle.shape
```

```python
# to change attacker1 to primary attacker

battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
# in place = true means the change will directly happen to data frame itself
```

```python
battle.head()
```

```python
#rename defender_1 to primary_defender

battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
```

```python
battle.head()
```

```python
#to know number times someone attacked some one

battle['attacker_king'].value_counts()
```

```python
battle['location'].value_counts()
```

```python
#pip install seaborn
# pip install matplotlib
# for visualization
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data = battle)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data = battle)
plt.show()
```

```python
sns.countplot(x= battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
